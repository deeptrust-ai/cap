```python
import os

import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
def create_heatmap(data):
    end = len(data) - 1
    mid = int(end/2)
    xticks = [0, mid, end]
    data = _preprocess(data)

    plt.imshow(data, cmap='autumn_r', interpolation='nearest')
    plt.colorbar(label="Deepfake Speech Likelihood")
    # plt.title("Deepfake Speech Heatmap\n(3 sec segments)")
    plt.title("Deepfake Speech Heatmap")
    plt.xlabel("Video Timeline (in seconds)")
    plt.xticks(xticks, [_to_date_string(x) for x in xticks])
    plt.yticks([])
    plt.show()

def _to_date_string(index):
    return str(datetime.timedelta(seconds=index * 3))
def _create_ticks(data):
    length = len(data) - 1
    

def delete_heatmap():
    os.remove("heatmap.png")

def _preprocess(data):
    data = np.array(data, dtype=np.float32)
    data = np.expand_dims(data, axis=0)

    return data
```
